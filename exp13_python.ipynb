```python
# preprocessing on a dataset-1
import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023",
"07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
df
```

```python
df.replace("-", np.nan, inplace=True)  # (replace "-" with NaN)
df
```

```python
# convert columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")
df
```

```python
# fill missing age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].median(), inplace=True)
df
```

```python
# fix inconsistent text formtting
df["Department"] = df["Department"].str.upper()
df
```

```python
# convert addmission_date to standard datetime format
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors="coerce")
df
# if a value can't be converted to a valid date, pandas will replace it with NaT (not a time)
# example: invalid dates like "abc" or 32-13-2024 become NaT
```

```python
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], format="mixed", dayfirst=True)
display(df)
```
